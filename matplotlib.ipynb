```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure(figsize=(4,3), dpi=100)
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='r')
ax.set_title('plot', fontsize=20)
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('y', fontsize=15)
```

```python
a = np.array([1,2,3,4,5])
b = np.array([8,1,5,1,8])
```

```python
plt.plot(a, b, color='b')
plt.show()
```

```python
x = np.arange(10)
y = np.arange(10)

fig, ax = plt.subplots(2)
for i in ax:
    i.plot(x, y, color='cyan')
```

```python
x = np.arange(10)
y = np.arange(10)

fig, ax = plt.subplots(2)
fig.set_figwidth(10)
fig.set_figheight(10)

for i in ax:
    i.plot(x, y, color='cyan')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

for i in ax:
    for j in i:
        j.plot(a, b, color='purple')
        j.set_title('plot')
        j.set_xlabel('x')
        j.set_ylabel('y')
fig.tight_layout()
plt.show()
```

```python
color = ['black', 'red', 'blue', 'orange', 'yellow', 'purple', 'cyan', 'green']
s = range(1, len(color)+1)
```

```python
for i ,j in zip(s, color):
    plt.plot([0, 10], [0, i], label=j, lw=3, color=j)
plt.legend(loc='upper left')
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
color = np.random.rand(N)

plt.scatter(x, y, c=color)
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
color = np.random.rand(N)

area = (30*np.random.rand(N))**2
plt.scatter(x, y, s=area, c=color, alpha=0.5)
```

```python
fig,ax=plt.subplots(nrows=2,ncols=3,figsize=(10,10))
s=np.random.rand(30,2)
ax[0][0].scatter(s[:,0],s[:,1],edgecolor='black',color='red',label='L=r')

ax[0][1].scatter(s[:,0],s[:,1],edgecolor='yellow',color='black',label='L=b')

ax[0][2].scatter(s[:,0],s[:,1],edgecolor='magenta',color='1',label='L=w_m')

ax[1][0].scatter(s[:,0],s[:,1],color='blue',label='L=b')

ax[1][1].scatter(s[:,0],s[:,1],color='green',label='L=g')

ax[1][2].scatter(s[:,0],s[:,1],edgecolor='cyan',color='1',label='L=w_c')

for i in ax:
    for j in i:
        j.legend(loc='upper right')

fig.tight_layout()
plt.show()
```

```python
lines=['-.', '--', 'None', '-', ':']
s = range(len(lines))
plt.figure(figsize=(8,4))

for i in s:
    plt.plot([i-1,i,i+1], [i,i,i],
             label="%s"%lines[i],
             linestyle=lines[i],
             lw=4)
plt.legend(loc='lower right')
plt.show()
```

```python
data=[np.random.normal(0,std,100) for std in range(1,5)]
fig=plt.figure(figsize=(8,6))
plt.boxplot(data,notch=True,sym='rs',vert=False)
plt.xticks([y+1 for y in range(len(data))],['x1','x2','x3','x4'])
plt.show()
```

```python
sns.set_style('darkgrid')
sns.boxplot(data)
```

```python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(10,5))
ax[0].violinplot(data,showmeans=False,showmedians=True)
ax[1].violinplot(data,showmeans=True,showmedians=True)
plt.show()
```
